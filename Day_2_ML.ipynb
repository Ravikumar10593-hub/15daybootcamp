# Data visulaization using matplotlib and seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#iris data set 

data=sns.load_dataset("iris")
data.head(3)
```

```python
data.shape
```

```python
#matplot

#line graph
plt.plot(data['sepal_width'])
plt.show()
```

```python
plt.plot(data['sepal_length'])
plt.show()
```

```python
plt.plot(data['petal_width'])
plt.show()
```

```python
#let's give name to x axis and y axis
plt.plot(data['sepal_width'])
plt.xlabel("Indexes")
plt.ylabel("petal length")
plt.show()
```

```python
#title to the graph,color,fontsize

plt.plot(data['sepal_width'],color="red")
plt.xlabel("Indexes")
plt.ylabel("petal length")
plt.title("sepal width graph",fontsize=27)
plt.show()
```

```python
#legend 
plt.plot(data['sepal_length'],color="red",label="sepal-l")
plt.plot(data['sepal_width'],color="green",label="sepal-w")
plt.xlabel("length")
plt.ylabel("width")
plt.title("Iris dataset",fontsize=27)
plt.legend()
plt.show()
```

```python
#figure size
plt.figure(figsize=(10,5))
plt.plot(data['sepal_length'],color="red",label="sepal-l")
plt.plot(data['sepal_width'],color="green",label="sepal-w")
plt.xlabel("length")
plt.ylabel("width")
plt.title("Iris dataset",fontsize=27)
plt.legend()
plt.show()
```

```python
#linestyle eg dotted
#marker  eg star at edges

plt.plot(data['sepal_length'],color="red",label="sepal-l",linestyle="--")
plt.plot(data['sepal_width'],color="green",label="sepal-w",marker="*")
plt.xlabel("length")
plt.ylabel("width")
plt.title("Iris dataset",fontsize=27)
plt.legend()
plt.show()
```

```python
#linewidth

plt.plot(data['sepal_length'],color="red",label="sepal-l",linestyle="--",linewidth=5)
plt.plot(data['sepal_width'],color="green",label="sepal-w",marker="*")
plt.xlabel("length")
plt.ylabel("width")
plt.title("Iris dataset",fontsize=27)
plt.legend()
plt.show()
```

```python
#grids
plt.plot(data['sepal_length'],color="red",label="sepal-l",linestyle="--")
plt.plot(data['sepal_width'],color="green",label="sepal-w",marker="*")
plt.xlabel("length")
plt.grid()
plt.ylabel("width")
plt.title("Iris dataset",fontsize=27)
plt.legend()
plt.show()
```
