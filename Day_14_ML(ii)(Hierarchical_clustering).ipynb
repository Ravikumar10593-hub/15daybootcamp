```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the csv file
data=pd.read_csv('/content/Mall_Customers.csv')
```

```python
#get indendent variable are used for cluster
x=data.iloc[:,[3,4]].values
```

```python
#implement HC and find the optimal number of cluster

import scipy.cluster.hierarchy as sch

dendrogram=sch.dendrogram(sch.linkage(x,method='ward'))

plt.title("Dendrogram")
plt.xlabel("Clusters")
plt.ylabel("Distance")
plt.show()
```

```python
#train the cluster with HC

from sklearn.cluster import AgglomerativeClustering

hc= AgglomerativeClustering(n_clusters=5,affinity="euclidean",linkage='ward')

y_hc=hc.fit_predict(x)
```

```python
#0-salary 1-scoring


plt.scatter(x[y_hc==0,0],x[y_hc==0,1],s=10,c='red',label="cluster1")
plt.scatter(x[y_hc==1,0],x[y_hc==1,1],s=10,c='orange',label="cluster2")
plt.scatter(x[y_hc==2,0],x[y_hc==2,1],s=10,c='yellow',label="cluster3")
plt.scatter(x[y_hc==3,0],x[y_hc==3,1],s=10,c='green',label="cluster4")
plt.scatter(x[y_hc==4,0],x[y_hc==4,1],s=10,c='blue',label="cluster5")


plt.title("Hc with agglomerative principle")
plt.legend()
plt.xlabel("annual income")
plt.ylabel("spending score")
plt.show()
```
